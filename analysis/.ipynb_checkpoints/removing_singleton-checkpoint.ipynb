```python
import dendropy
import pandas as pd 
import baltic as bt
```

```python
trees = "/Users/miguelparedes/Desktop/glm_multi/rhino/variant_builds/results/glm_randomkc_clusters_combined_3000.typed.trees"
```

```python
tree_list = dendropy.TreeList()
tree_list.read(path=trees, schema="nexus", extract_comment_metadata=True, ignore_unrecognized_keyword_arguments=False)
```

```python
tree_list
```

```python
for index, tree in enumerate(tree_list):
    tree.print_plot()
    if index >0:
        break 
```

```python
for tree in tree_list:
    for edge in tree.postorder_edge_iter():
        if edge.head_node.num_child_nodes() == 1:
            edge.collapse()
        
```

```python
dest = open("randomkc_trees_noscnode_3000.trees", "w")
tree_list.write(file=dest, schema="nexus")
```
