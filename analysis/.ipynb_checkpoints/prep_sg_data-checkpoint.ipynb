```python
ne_df = dfnew[(dfnew.category == "within North") | (dfnew.category == "within South")]
#ne_df = ne_df[ne_df.start_date > "2019-12-10"]
#ne_df = ne_df[ne_df.start_date <= "2021-10-18"]
```

```python
ne_df.start_date = pd.to_datetime(ne_df['start_date'], format='%Y-%m-%d')
```

```python
ne_df = ne_df.sort_values(by=['start_date', 'category'], ascending = [True, False])
```

```python
ne_df
```

```python
first_date = pd.Timestamp('2019-12-15')
end_date = pd.Timestamp('2022-02-27')
```

```python
ne_df = ne_df.set_index("start_date")
```

```python
nne_df = ne_df[ne_df.category == "within North"]
sne_df = ne_df[ne_df.category == "within South"]
```

```python
nne_df
```

```python
ne_df
```

```python
dates = pd.date_range(first_date, end_date, freq='D')
dates.name = 'date'
```

```python
n_daily_df = nne_df.reindex(dates, method='ffill')
s_daily_df = sne_df.reindex(dates, method = 'ffill')
```

```python
n_daily_df = n_daily_df.drop(["poi_geo", "visitor_geo", "category"], axis = 1)
s_daily_df = s_daily_df.drop(["poi_geo", "visitor_geo", "category"], axis = 1)
```

```python
n_daily_df
```

```python
sn_merged_ne_df = pd.merge(s_daily_df, n_daily_df, left_index=True, right_index=True)
```
