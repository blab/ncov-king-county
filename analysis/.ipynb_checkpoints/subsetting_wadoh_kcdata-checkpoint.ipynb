```python
u_df =df[df['ID'].duplicated() == False]
```

```python
u_df[u_df['POSTAL_CODE'] == 'USA'] = np.nan
u_df[u_df['POSTAL_CODE'] == 'undef-ined'] = np.nan
u_df[u_df['POSTAL_CODE'] == 'UNDEF'] = np.nan
u_df = u_df.dropna(subset=['POSTAL_CODE'])
```
