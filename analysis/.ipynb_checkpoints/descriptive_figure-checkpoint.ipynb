```python
def convert_format_year(date):
    #date = dt.strptime(number, '%Y-%m-%d')
    date_new = dt.strftime(date, '%Y')
    return date_new
```

```python
def convert_format_month(date):
    #date = dt.strptime(number, '%Y-%m-%d')
    date_new = dt.strftime(date, '%m')
    return date_new
```

```python
north_df["year"] = north_df['start_date'].map(convert_format_year)
south_df["year"] = south_df['start_date'].map(convert_format_year)
```

```python
north_df["month"] = north_df['start_date'].map(convert_format_month)
south_df["month"] = south_df['start_date'].map(convert_format_month)
```

```python
north_df
```

```python
north_df.set_index(['month','year'], inplace=True)
south_df.set_index(['month','year'], inplace=True)
```

```python
north_df
```

```python
df1_n = north_df.pivot_table(values='sum_visits', index='month', columns='year')
df1_s = south_df.pivot_table(values='sum_visits', index='month', columns='year')
```

```python
df1_s
```

```python
df1_n["20-19"] = (df1_n['2020'] - df1_n['2019'])/df1_n['2019']
df1_s["20-19"] = (df1_s['2020'] - df1_s['2019'])/df1_s['2019']
```

```python
df1_n["21-19"] = (df1_n['2021'] - df1_n['2019'])/df1_n['2019']
df1_s["21-19"] = (df1_s['2021'] - df1_s['2019'])/df1_s['2019']
```

```python
df1_n = df1_n[['20-19', '21-19']]
df1_s = df1_s[['20-19', '21-19']]
```

```python
df1_n = df1_n.reset_index()
df1_s = df1_s.reset_index()
```

```python
df1_n
```
