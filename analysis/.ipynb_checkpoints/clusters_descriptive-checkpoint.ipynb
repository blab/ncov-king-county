```python
new_clus =  "/Users/miguelparedes/Desktop/gitrepos/ncov-king-county/Simulations/Validation-Constant/data/kc_clusters.tsv"
```

```python
clus_df = pd.read_csv(new_clus, sep="\t", parse_dates=  [2])
```

```python
clus_df
```

```python
cluster_counts_n = pd.DataFrame()
cluster_counts_s = pd.DataFrame()
cluster_counts_n['count_clus'] = clus_df[clus_df['location'] =="North_King_County"].cluster.value_counts()
cluster_counts_s['count_clus'] = clus_df[clus_df['location'] =="South_King_County"].cluster.value_counts()
```

```python
cluster_counts_n = cluster_counts_n.reset_index()
cluster_counts_n = cluster_counts_n.rename(columns={"index": "clusid"})
cluster_counts_s = cluster_counts_s.reset_index()
cluster_counts_s = cluster_counts_s.rename(columns={"index": "clusid"})
```

```python
cluster_counts_n
```

```python
count_n =  alt.Chart(cluster_counts_n).mark_bar(opacity = 1).encode(
    x=alt.X('count_clus:O', bin = alt.Bin(maxbins=100,step=10), axis=alt.Axis(title="Cluster Size", grid=False)),
    y=alt.Y('count()',  axis=alt.Axis(title="Number of Local Outbreak Clusters", grid=False)))
 
count_s =  alt.Chart(cluster_counts_s).mark_bar(opacity = 1,color = "orange").encode(
    x=alt.X('count_clus:O', bin = alt.Bin(maxbins=100,step=10), axis=alt.Axis(title="Cluster Size", grid=False)),
    y=alt.Y('count()',  axis=alt.Axis(title="Number of Local Outbreak Clusters", grid=False)))
    
```

```python
cluster_count = count_n | count_s
```

```python
cluster_count
```

```python
cluster_count.save("cluster_count.png")
```

```python
large_clus_n = cluster_counts_n[cluster_counts_n.count_clus > 10]
large_clus_s = cluster_counts_s[cluster_counts_s.count_clus > 10]
```

```python
large_count_n =  alt.Chart(large_clus_n).mark_bar().encode(
    x=alt.X('count_clus:O', bin = alt.Bin(maxbins=20,step=10), axis=alt.Axis(title="Cluster Size", grid=False)),
    y=alt.Y('count()', axis=alt.Axis(title="Number of Local Outbreak Clusters", grid=False), scale=alt.Scale(domain=(0, 35))))

large_count_s =  alt.Chart(large_clus_s).mark_bar(color = "orange").encode(
    x=alt.X('count_clus:O', bin = alt.Bin(maxbins=20,step=10), axis=alt.Axis(title="Cluster Size", grid=False)),
    y=alt.Y('count()', axis=alt.Axis(title="Number of Local Outbreak Clusters", grid=False), scale=alt.Scale(domain=(0, 35))))
```

```python
large_clus_count = large_count_n | large_count_s
```

```python
large_clus_count
```

```python
large_clus_count.save("large_clus_count.png")
```

```python
combined_clus_count = cluster_count & large_clus_count
```

```python
combined_clus_count.save("combined_clus_count.png")
```

```python
date_clus = clus_df.groupby('cluster').first().reset_index()
```

```python
date_clus
```

```python
def convert_format(number):
    date = dt.strptime(number, '%Y-%m-%d')
    date = dt.strftime(date, '%Y-%m')
    return date
```

```python
date_clus['year-month'] = date_clus['date'].map(convert_format)
```

```python
date_clus
```

```python
clust_over_time =  alt.Chart(date_clus).mark_bar(opacity = 1, interpolate='monotone').encode(
    alt.X('year-month:O'),
    alt.Y('count():Q', axis=alt.Axis(title="Number of Local Outbreak Clusters", grid=False), stack = None,), 
    alt.Color("location"), 
    column="location:N")
```

```python
clust_over_time 
```

```python
clust_over_time.save("clust_over_time.png")
```
