```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
```

```python
def convert_format(number):
   # date = dt.strptime(number, '%m/%d/%Y')
    date = dt.strftime(number, '%Y%U')
    return date
```

```python
def take_first(array_like):
    return array_like[0]
```

```python
hosps = '/Users/miguelparedes/Downloads/overall_daily-may-30.csv'#data taken from https://kingcounty.gov/depts/health/covid-19/data/daily-summary.aspx
```

```python
hosp_df = pd.read_csv(hosps, parse_dates = [0])
```

```python
hosp_df
```

```python
hosp_df['yearweek'] = hosp_df.Admission_Date.map(convert_format)

```

```python
hosp_df
```

```python
hosp_df = hosp_df.set_index('Admission_Date')
```

```python
hosp_df = hosp_df.resample('W').agg({'Hospitalizations': "sum", 
                               'Hospitalization_Rate': 'sum', 
                                "yearweek": take_first})
```

```python
hosp_df.to_csv("kc_hosp_weekly.csv")
```
