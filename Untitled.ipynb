```python
import pandas as pd
import altair as alt
import numpy as np
from altair import datum
```

```python
raw_metadata = pd.read_csv("data/sequenced_kingco_3.22.22.csv")
alpha_clusters = pd.read_csv("mascot_glm/data/kc_clusters_alpha_new.tsv", sep = "\t")
```

```python
alpha_clusters
```

```python
raw_metadata
```

```python
merged_df = pd.merge(alpha_clusters, raw_metadata, left_on= "strain", right_on = "GISAID_ID")
merged_df
```

```python
merged_df["kids"] = np.where(((merged_df.AGE_DECADE == "Age 0-9") | (merged_df.AGE_DECADE == "Age 10-19")), "children", "adult")
merged_df["older"] = np.where(((merged_df.AGE_DECADE == "Age 60-69") | (merged_df.AGE_DECADE == "Age 70-79") | (merged_df.AGE_DECADE == "Age 80-89") |(merged_df.AGE_DECADE == "Age 90+") ), "elderly", "adult")
merged_df
```

```python
merged_df['year_month']=pd.to_datetime(merged_df['date']).dt.strftime('%y-%m')
merged_df['year_week']=pd.to_datetime(merged_df['date']).dt.strftime('%y-%U')

merged_df
```

```python
cluster_counts_old = pd.DataFrame()
cluster_counts_adult = pd.DataFrame()
cluster_counts_old['count_clus'] = merged_df[merged_df['older'] =="elderly"].cluster.value_counts()
cluster_counts_adult['count_clus'] = merged_df[merged_df['older'] =="adult"].cluster.value_counts()
cluster_counts_old = cluster_counts_old.reset_index()
cluster_counts_old = cluster_counts_old.rename(columns={"index": "clusid", "count_clus": "size", "older": "age"})
cluster_counts_adult = cluster_counts_adult.reset_index()
cluster_counts_adult = cluster_counts_adult.rename(columns={"index": "clusid", "count_clus": "size", "older": "age"})

cluster_counts_old["age"] = "older"
cluster_counts_adult["age"] = "adult"
```

```python
concat_clus = pd.concat([ cluster_counts_old, cluster_counts_adult], ignore_index=True)
concat_clus = concat_clus[["size", "age"]]
```

```python
concat_clus
```

```python
concat_clus.groupby("age")["size"].mean()
```

```python
cluster_size = merged_df["cluster"].value_counts().reset_index().rename(columns={"index": "cluster", "cluster": "size"})
```

```python
elderly_count = merged_df.groupby("cluster")["older"].value_counts().reset_index(level=0, drop=False).rename(columns={"older": "count"}).reset_index()
```

```python
combined_elderly_count = pd.merge(cluster_size, elderly_count, on = "cluster")
combined_elderly_count = combined_elderly_count[combined_elderly_count.older == "elderly"]
```

```python
combined_elderly_count["prop_elder"] = combined_elderly_count["count"]/combined_elderly_count["size"]
```

```python
combined_elderly_count[combined_elderly_count["size"] > 5]["prop_elder"].mean()
```
